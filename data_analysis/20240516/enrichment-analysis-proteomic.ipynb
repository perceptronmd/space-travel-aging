# SOMA aging tissues (mouse) with top pro-aging proteins from PROTEOMICS model

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Creating DataFrame
df = pd.read_csv("./SOMA_PROTEOMICS_OLDER_MOUSE_RAW.csv")
# Converting #NUM! and #DIV/0! to NaN
df["enrichment_score"] = df["enrichment_score"].replace({"#NUM!": np.nan, "#DIV/0!": np.nan}).astype(float)
df["OR"] = df["OR"].replace({"#NUM!": np.nan, "#DIV/0!": np.nan}).astype(float)
df["-log10P"] = df["-log10P"].replace({"#NUM!": np.nan, "#DIV/0!": np.nan}).astype(float)

# add enrichment scores that preserve direction of the effect
df["enrichment_score_w_direction"] = (1-df["OR"])/df["padj"]


# Calculating the average enrichment score for each tissue type
OLDER_average_enrichment = df.groupby("TissueType")["enrichment_score"].mean().reset_index()
OLDER_average_enrichment['enrichment_score_log10'] = np.log10(OLDER_average_enrichment['enrichment_score'])
OLDER_average_enrichment_w_direction= df.groupby("TissueType")["enrichment_score_w_direction"].mean().reset_index()
OLDER_average_enrichment_w_direction['direction'] = [1 if x>0 else -1 for x in OLDER_average_enrichment_w_direction['enrichment_score_w_direction']]
OLDER_average_enrichment_w_direction['enrichment_score_log10'] = np.log10(np.abs(OLDER_average_enrichment_w_direction['enrichment_score_w_direction']))*OLDER_average_enrichment_w_direction['direction']



# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(OLDER_average_enrichment["TissueType"], OLDER_average_enrichment["enrichment_score_log10"], color='red')
plt.xlabel("Tissue Type")
plt.ylabel("Average Enrichment Score (log10)")
plt.title("Average Enrichment Score Across Each Tissue Type")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(OLDER_average_enrichment_w_direction["TissueType"], OLDER_average_enrichment_w_direction["enrichment_score_log10"], color='red')
plt.xlabel("Tissue Type")
plt.ylabel("(Average Enrichment Score (log10)\n <-- younger | older -->")
plt.title("Pro-aging proteins analysis")
plt.xticks(rotation=45)
plt.show()
```
